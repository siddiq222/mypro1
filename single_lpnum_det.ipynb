```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
import easyocr
reader=easyocr.Reader(['en'])
image=cv2.imread(r"C:\Users\hussain siddiq\Downloads\lp2.png")
color_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
print(color_img.shape)
plt.imshow(color_img)
model=YOLO(r"C:\Users\hussain siddiq\Downloads\license_plate_detector.pt")
result=model.predict(color_img,save=True)
print(result[0].boxes.xyxy.numpy()[0])
x1=int(result[0].boxes.xyxy.numpy()[0][0])
y1=int(result[0].boxes.xyxy.numpy()[0][1])
x2=int(result[0].boxes.xyxy.numpy()[0][2])
y2=int(result[0].boxes.xyxy.numpy()[0][3])
print(x1,y1,x2,y2)
plt.imshow(color_img[y1:y2,x1:x2])
lpnum=reader.readtext(color_img[y1:y2,x1:x2])
print(lpnum[0][1])
```
